```python
colors =['magenta','green', 'blue']
species =['Iris-virginica','Iris-versicolor', 'Iris-setosa']
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLength(cm)'], x['SepalWidth(cm)'], c = colors[i], label=species[i])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.legend()
```

```python
for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['PetalLength(cm)'], x['PetalWidth(cm)'], c = colors[i], label=species[i])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.legend()
```

```python
# Model Training
from sklearn.model_selection import train_test_split
# train - 70
# test - 30
X = df.drop(columns=['Species'])
Y= df['Species']
x_train, x_test, y_train, y_test= train_test_split(X, Y, test_size=0.30)
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
#to make prediction
import numpy as np
x_new=np.array([[5,2.9,1,0.2]])
x_new.shape
```

```python
prediction=model.predict(x_new)        #Therefore x_new belongs to iris setosa
prediction
```

```python
# print metrix to get performance
print("Accuracy: ",model.score(x_test, y_test)*100)
```

```python
# knn - k-nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(x_train, y_train)
```

```python
#to make prediction
import numpy as np
x_new=np.array([[5,2.9,1,0.2]])
x_new.shape
```

```python
prediction=model.predict(x_new)        #Therefore x_new belongs to iris setosa
prediction
```

```python
# print metrix to get performance
print("Accuracy: ",model.score(x_test, y_test)*100)
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
# print metrix to get performance
print("Accuracy: ",model.score(x_test, y_test)*100)
```
