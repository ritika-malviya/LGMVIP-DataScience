# Importing libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
# Loading the image 
```

```python
img = cv2.imread("D:/Data/Downloads/images.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

```python
 # Converting Image to GrayScale
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
# Inverting the Image
```

```python
img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize=(8,8))
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.show()
```

```python
# smoothening image
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21,21),sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smoothing,cmap="gray")
plt.axis("off")
plt.title("Smoothen Image")
plt.show()
```

```python
 # Converting your images to pencil sketches
```

```python
final = cv2.divide(img_gray, 255 - img_smoothing, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```

```python
# The Final Output
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,2,2)
plt.imshow(final,cmap="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
